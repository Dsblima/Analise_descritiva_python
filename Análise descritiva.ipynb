```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
import math
import numpy as np

trab = pd.read_csv("C:\\Users\\danil\\Documents\\Mestrado\\Mineração\\Analise Descritiva\\german_credit_data.csv").drop(columns=['id'])
print('Descrição')
print (trab.describe())
print('\n')
trab.head()
print('\n')
print("INFO")
trab.info()
print('\n')
print("CORRELAÇÃO")
print(trab.corr())
print('\n')
print("COVARIÂNCIA")
print(trab.cov())
```

```python
print("Idade")
print("Moda")
print(trab['Age'].mode())
print("Amplitude")
print(trab['Age'].max() - trab['Age'].min())
print("Variância")
print(trab['Age'].var())
```

```python
print("Quantidade de Crédito")
print("Moda")
print(trab['Credit amount'].mode())
print("Amplitude")
print(trab['Credit amount'].max() - trab['Credit amount'].min())
print("Variância")
print(trab['Credit amount'].var())
```
